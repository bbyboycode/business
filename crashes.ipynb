```python
import pandas as pd
```

```python
crashes = pd.read_csv('crashes.csv')
```

```python
crashes
```

```python
# drop all rows that have NaN for latitude or longitude
crashes.dropna(subset = ['LATITUDE', 'LONGITUDE'], inplace = True)
```

```python
# drop all rows that have NaN for latitude or longitude
crashes.dropna(subset = ['BOROUGH'], inplace = True)
```

```python
# convert CRASH DATE to datetime format
crashes['CRASH DATE'] = pd.to_datetime(crashes['CRASH DATE'])
```

```python
crashes['year'] = crashes['CRASH DATE'].dt.year
crashes
```

```python
crashes.info()
```

```python
crashes2018 = crashes.loc[crashes['year'] == 2018]
```

```python
crashes_samp2018 = crashes2018.sample(n = 750)
```

```python
crashes2019 = crashes.loc[crashes['year'] == 2019]
```

```python
crashes_samp2019 = crashes2019.sample(n = 750)
```

```python
crashes2020 = crashes.loc[crashes['year'] == 2020]
```

```python
crashes_samp2020 = crashes2020.sample(n = 750)
```

```python
crashes2021 = crashes.loc[crashes['year'] == 2021]
```

```python
crashes_samp2021 = crashes2021.sample(n = 750)
```

```python
crashes_samp = pd.concat([crashes_samp2018,
                          crashes_samp2019,
                          crashes_samp2020,
                          crashes_samp2021],
                          axis = 0)
```

```python
crashes_samp.to_csv('crashes_cleaned_equal_samp.csv')
```
